```python
class shifted_ackley:
    def __init__(self, dims):
        self.dims = dims
    def fitness(self, x):
        bias = -140.0
        z = np.array(x) - np.array(ackley[:self.dims])
        S1 = np.sqrt(np.sum(np.square(z)) / self.dims)
        S2 = np.exp(np.sum(np.cos(2*np.pi*z)) / self.dims)
        global fitness_results
        val = (-20*np.exp(-0.2*S1-S2)) + 20 + np.exp(1) + bias
        fitness_results.append(val)
        return [val, ]
    
    
    def get_bounds(self):
        return ([-32]*self.dims,[32]*self.dims)
    
def execute_algorithm(dims):
    global fitness_results
    fitness_results = []
    current_algorithm = None
    current_problem = None
    if(dims == 50):
        current_algorithm = algorithm(pso(gen = 5000,  omega=0.7, eta1=4, eta2=.5))
        current_problem = pg.problem(shifted_ackley(50))
    elif(dims == 500):
        current_algorithm = algorithm(simulated_annealing(Ts=1, Tf=1e-12, n_T_adj=15, n_range_adj=5, bin_size=5, start_range=1))
        current_problem = pg.problem(shifted_ackley(500))
    else : 
        return "Provide correct dimensions"
        

    current_population = population(current_problem, population_size)
    start_time = time.time()
    current_population_result = current_algorithm.evolve(current_population)
    ended_time = time.time()
    completed_time = ended_time - start_time
    print(current_population_result)
    print("\nComputational time:", round(completed_time, 2), "seconds\n")
```
